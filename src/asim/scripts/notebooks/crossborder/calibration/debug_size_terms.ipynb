```python
ctramp_taz_st = mgra.groupby('taz')['size_term'].sum()
```

```python
presample_alts = pd.read_csv('output/trace.tour_od_choice.sample.cargo.presample.interaction_sample.chunk_1.eval.household_id.79.csv')
```

```python
presample_trace_alts = presample_alts.set_index('expression').T
```

```python
presample_trace_alts['ctramp_st'] = ctramp_taz_st.reindex(presample_trace_alts['dest_TAZ'].astype(int)).values
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(presample_trace_alts['size_term'].astype(float), presample_trace_alts['ctramp_st'])

ax.set_xlabel('tour_od_choice trace alts')
ax.set_ylabel('manually computed')
ax.set_title('presample size terms (TAZ)')
```
