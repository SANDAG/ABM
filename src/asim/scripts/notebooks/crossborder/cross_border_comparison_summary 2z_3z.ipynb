```python
asim_tour_trip_modes = asim_trips.groupby('tour_id')['trip_mode'].unique()
asim_tours['tour_trip_modes'] = asim_tour_trip_modes.reindex(asim_tours['tour_id']).values
asim_tours['tour_mode_in_trips'] = asim_tours.apply(lambda x: x['tour_mode'] in x['tour_trip_modes'], axis=1)
```

```python
Survey_tour_trip_modes = Survey_trips.groupby('tour_id')['trip_mode'].unique()
Survey_tours['tour_trip_modes'] = Survey_tour_trip_modes.reindex(Survey_tours['tour_id']).values
Survey_tours['tour_mode_in_trips'] = Survey_tours.apply(lambda x: x['tour_mode'] in x['tour_trip_modes'], axis=1)
```

```python
asim_tour_mode_in_trips = asim_tours.groupby('tour_mode').agg({'tour_mode_in_trips':'value_counts'})
asim_tour_mode_in_trips.columns = ['count']
asim_tour_mode_in_trips['source'] = 'asim'

asim_tour_mode_in_trips['total'] = asim_tour_mode_in_trips.groupby('tour_mode')['count'].sum().reindex(
    asim_tour_mode_in_trips.index.get_level_values(0).values).values

asim_tour_mode_in_trips['pct'] = asim_tour_mode_in_trips['count'] / asim_tour_mode_in_trips['total']
```

```python
Survey_tour_mode_in_trips = Survey_tours.groupby('tour_mode').agg({'tour_mode_in_trips':'value_counts'})
Survey_tour_mode_in_trips.columns = ['count']
Survey_tour_mode_in_trips['source'] = 'Survey'

Survey_tour_mode_in_trips['total'] = Survey_tour_mode_in_trips.groupby('tour_mode')['count'].sum().reindex(
    Survey_tour_mode_in_trips.index.get_level_values(0).values).values

Survey_tour_mode_in_trips['pct'] = Survey_tour_mode_in_trips['count'] / Survey_tour_mode_in_trips['total']
```

```python
tour_mode_in_trips = pd.concat((asim_tour_mode_in_trips,Survey_tour_mode_in_trips),axis=0)
tour_mode_in_trips.reset_index(inplace=True)
```

```python
tour_mode_in_trips = tour_mode_in_trips.loc[tour_mode_in_trips['tour_mode_in_trips'] == True,['tour_mode','source','pct']]
```
